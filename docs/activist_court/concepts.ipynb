## Types of Modelled Conflicts

| Term                | Definition                                                       | Mark |
| ---                 | ---                                                              | ---  |
| **Contradiction**   | neglected or refused to state apparent inconsistency or conflict | '?x' |
| **Deception**       | intentionally emphasized relevant half-truths to affect judgment | '?d' |
| **Distortion**      | reported relevant information in a suggestive and biased manner  | '?s' |
| **Fairness**        | neglected or refused to consider both claim and counterclaim     | '?a' |

## Types of Modelled Judgments

| Term            | Definition                                                         | Mark |
| ---             | ---                                                                | ---  |
| **Bias**        | selection and judgment of conflicts satisfying a prior narrative   | '=b' |
| **Confusion**   | selection and judgment of conflicts leading to clear contradiction | '=c' |
| **Disregard**   | avoidance of relevant conflicts while protecting a prior narrative | '=g' |
| **Fabrication** | injection of disputable conjecture reinforcing a prior narrative   | '=f' |
| **Validation**  | selection and judgment of conflicts refuting an inflated Conflict  | '=v' |
