# A Conceptual Model Of An "Emotional" Domain

We are only concerned with textual data stored electronically. The `source` is thus our unit of representation.

The `evidence` is all the `doc`s we need to consider organized hierarchically as `source`s.

Ultimately, it all comes down to working with text fragments, `claims`, with various interconnected and even overlapping properties. We define:

| Term         | Definition                             | Properties                               |
| ---          | ---                                    | ---                                      |
| **Source**   | digitized content stored in a file     | type (e.g. pdf, jpg, email, text)        |
| **Evidence** | a hierarchy of considered Sources      | root (directory)                         |
| **Doc**      | a unit of paginated flat text Evidence | name (date), genre, author, title, pages |
| **Genre**    | a kind of "officialness" of a Doc      | name                                     |
| **Place**    | a specific location in a Doc           | doc, page, para(graph)                   |
| **Claim**    | a Placed relevant text fragment        | text, place                              |

The `author` of a `doc` of `claim`s can be a party, a lawyer or a judge, for example. Accordingly, we define:

| Term          | Definition                            | Properties                           |
| ---           | ---                                   | ---                                  |
| **Author**    | an author of Docs of Claims           | name, title                          |
| **Agent**     | a to be believed professional         | kind (of profession)                 |
| **Authority** | a trusted Agent of an official agency | agency (institution or 'sworn self') |

A `claim` can be grouped into `topic`s and attached to `narrative`s. Such groupings and assignments are done based on the semantic meaning of the claim as well as the context the claim belongs to.

| Term          | Definition                                    | Properties  |
| ---           | ---                                           | ---         |
| **Topic**     | a relevant/dominant subject matter of Claims  | name        |
| **Narrative** | a thematically focused "slice" through Claims | name, topic |

More importantly, `claim`s are further categorized as follows. The objective is to either compute a generalizing `credibility` for the claim or to derive other reproducible conjecture regarding the related claims. 

| Term           | Definition                                                | Properties       |
| ---            | ---                                                       | ---              |
| **Node**       | a generic grouping of Claim(s) along thematic dimensions  | topic, narrative |
| **Proof**      | a readily verifiable undisputed Claim from an Authority   | claim, authority |
| **Conjecture** | a logical inference from Proofs guided by thematic Claims | proofs, claims   |
| **Reality**    | a broad-based reinforcing Conjecture                      | coherence        |
| **Dissent**    | a narrow, fragmenting, and thus weakening, Conjecture       | fragment         |

A `dissent` is further broken down into self-explained types with increasing factors of their fragmenting effects. These are:

| Term                | Definition                                                       |
| ---                 | ---                                                              |
| **Misreading**      | stated seemingly unintentionally misinterpreted relevant claims  |
| **Fairness**        | neglected or refused to consider both claim and counterclaim     |
| **Negligence**      | reported relevant and damaging information in a careless manner  |
| **Proportionality** | neglected or refused to report in balanced/proportional manner   |
| **Contradiction**   | neglected or refused to state apparent inconsistency or conflict |
| **Isolation**       | neglected or refused to consider relevant and essential context  |
| **Omission**        | neglected or refused to report disclosed relevant information    |
| **Distortion**      | reported relevant information in a suggestive and biased manner  |

A special type of a claim is a `conflict`. Its purpose is to directly contradict a `reality` and thus to manipulate a subsequent `judgment`.

| Term         | Definition                                                      | Properties           |
| ---          | ---                                                             | ---                  |
| **Conflict** | a Claim intended to fabricate/escalate bases for Judgments      | claim, reality       |
| **Suborn**   | a false Claim caused by Agent as intentional perjury            | (agent's) agency     |
| **Repeat**   | variant of conflict inducing Claim to reinforce intended effect | claim, conflict      |
| **Judgment** | a thematic conjecture by Authority over inflated Conflicts      | conflicts, authority |
| **Activism** | a thematic pattern of conjectures by Authority                  | judgments, authority |

Revealing, and even fabricating, conflicts is the objective of lawsuits. We categorize conflicts along their increasing fragmenting effect on judgments: 

| Term         | Definition                                                       |
| ---          | ---                                                              |
| **Inherent** | emphasized inherent contradictions to confuse judgment           |
| **Conceal**  | intentionally concealed relevant truths to affect judgment       |
| **Deceive**  | intentionally emphasized relevant half-truths to affect judgment |
| **Fraud**    | stated intentionally falsified relevant and crucial information  |
| **Extort**   | stated relevant and threatening claims to force money or "fees"  |

The judgments of lawsuits represent the net effect of all the claimed conflicts. If a `judgment` relies predominantly on conflicts while overruling realities, it causes `turmoil`. The types of judgments we consider in increasing turmoil effect is as follows: 

| Term            | Definition                                                         |
| ---             | ---                                                                |
| **Validation**  | selection and judgment of conflicts refuting an inflated Conflict  |
| **Confusion**   | selection and judgment of conflicts leading to clear contradiction |
| **Bias**        | selection and judgment of conflicts satisfying a prior narrative   |
| **Disregard**   | avoidance of relevant conflicts while protecting a prior narrative |
| **Fabrication** | injection of disputable conjecture reinforcing a prior narrative   |

When considering multiple judgment claims, a pattern, a trend can possibly be established. During processing, when our calculations support such patterns, `activism`s emerge. We define the following categories in increasing effect on their `turmoil` factor:
